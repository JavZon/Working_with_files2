```python
query="PRAGMA table_info(elect);"
cursor.execute(query)
cursor.fetchall()
```

```python
query="SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country='Uzbekistan' AND year BETWEEN 2005 and 2015"
```

```python
df=pd.read_sql(query, db)
df[(df['year']==2005) | (df['year']==2010) | (df['year']==2015)]
```
