```python
db=sql.connect('chinook.db')
cursor=db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table' ;")
cursor.fetchall()
```

```python
q1="SELECT COUNT(*) FROM albums "
cursor.execute(q1)
cursor.fetchall()
```

```python
cursor.execute("PRAGMA table_info(albums)")
cursor.fetchall()
```

```python
q2="SELECT * FROM albums LIMIT 10"
df=pd.read_sql(q2, db)
df
```

```python
q2="SELECT TrackId,Name,Composer,Bytes FROM tracks LIMIT 5"
df=pd.read_sql(q2, db, index_col='TrackId')
df
```

```python
q2="SELECT TrackId,Name,Composer,Bytes FROM tracks WHERE Composer IS NOT NULL ORDER BY Name DESC LIMIT 10"
df=pd.read_sql(q2, db, index_col='TrackId')
df
```

```python
q3="SELECT DISTINCT Composer FROM tracks WHERE Composer IS NOT NULL AND Composer LIKE 'S%'"
df2=pd.read_sql(q3, db)
df2.head()
```
