```python
import pandas as pd

# Step 1: Create a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 22],
        'City': ['New York', 'San Francisco', 'Los Angeles']}

df = pd.DataFrame(data)

# Step 2: Save the DataFrame to an HDF5 file
df.to_hdf('example.h5', key='my_hdf_file', mode='w')

# Step 3: Read the HDF5 file back into a DataFrame
read_df = pd.read_hdf('example.h5', key='my_hdf_file')

# Step 4: Display the original DataFrame and the one read from the HDF5 file
print("Original DataFrame:")
print(df)

print("\nDataFrame read from HDF5 file:")
print(read_df)
```
