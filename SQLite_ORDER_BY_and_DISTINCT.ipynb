```python
db=sql.connect('chinook.db')
cursor=db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type= 'table' ; ")
cursor.fetchall()
```

```python
query="SELECT * FROM customers LIMIT 10 "
df=pd.read_sql(query, db)
df
```

```python
query="SELECT CustomerId,FirstName,LastName,Country,Phone FROM customers LIMIT 10"
df=pd.read_sql(query, db, index_col='CustomerId')
df
```

```python
query="SELECT CustomerId,FirstName,LastName,Country,Phone FROM customers ORDER BY Country LIMIT 10"
df=pd.read_sql(query, db, index_col='CustomerId')
df
```

```python
query="SELECT CustomerId,FirstName,LastName,Country,Phone FROM customers ORDER BY Country DESC LIMIT 20"
df=pd.read_sql(query, db, index_col='CustomerId')
df
```

```python
q="SELECT DISTINCT Country FROM customers"
df=pd.read_sql(q, db)
df
```
