```python
import sqlite3 as sql
import pandas as pd
import numpy as np
```

```python
db=sql.connect('my_base.db')
cursor=db.cursor()
```

```python
dict1={
    'AS Roma':[2017,2023,2025],
    'FC Milan':[2013,2016,2020],
    'FC Porto':[1998,2012,2018]
    }
df=pd.DataFrame(dict1)
df
```

```python
df.to_sql("clubs", db)
```

```python
query="SELECT name FROM sqlite_master WHERE type='table';"
cursor.execute(query)
cursor.fetchall()
```

```python
query="PRAGMA table_info(clubs)"
cursor.execute(query)
cursor.fetchall()
```

```python
query="SELECT * FROM clubs"
cursor.execute(query)
cursor.fetchall()
```

```python
pd.read_sql("SELECT * FROM clubs", db, index_col='index')
```
