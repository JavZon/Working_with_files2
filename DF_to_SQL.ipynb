```python
columns = ['EmployeeID', 'First_Name', 'Last_Name', 'BornYear', 'Position', 'Salary', 'Phone']
data = {
    'EmployeeID': np.random.randint(1000, 9999, size=20),
    'First_Name': [np.random.choice(['John', 'Alice', 'Michael', 'Emma', 'David']) for _ in range(20)],
    'Last_Name': [np.random.choice(['Smith', 'Johnson', 'Williams', 'Brown', 'Jones']) for _ in range(20)],
    'BornYear': np.random.randint(1970, 2000, size=20),
    'Position': [np.random.choice(['Manager', 'Developer', 'Designer', 'Analyst','IT Staff']) for _ in range(20)],
    'Salary': np.random.randint(30000, 100000, size=20),
    'Phone': [''.join(np.random.choice([str(num) for num in range(10)], size=10)) for _ in range(20)],

}
```

```python
df = pd.DataFrame(data, columns=columns)
df
```

```python
db=sql.connect('example.db')
cursor=db.cursor()
```

```python
df.head()
```

```python
df.to_sql('employees',db)
```

```python
query="SELECT name FROM sqlite_master WHERE type='table'"
cursor.execute(query)
cursor.fetchall()
```

```python
q2="SELECT COUNT(*) FROM employees"
cursor.execute(q2)
cursor.fetchall()
```

```python
q2="PRAGMA table_info(employees)"
cursor.execute(q2)
cursor.fetchall()
```

```python
query="SELECT * FROM employees WHERE BornYear>1980 AND Position ='IT Staff' ORDER by BornYear "
df=pd.read_sql(query, db, index_col='index')
df
```
