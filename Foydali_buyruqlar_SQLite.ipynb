```python
db=sql.connect('electricity.db')
cursor=db.cursor()
```

```python
query="SELECT name FROM sqlite_master WHERE type='table'"
cursor.execute(query)
cursor.fetchall()
```

```python
query="PRAGMA table_info(elect)"
cursor.execute(query)
cursor.fetchall()
```

```python
df=pd.read_sql("SELECT * FROM elect ",db,index_col='index')
df['country'].value_counts().nlargest(3)
```

```python
df=pd.read_sql("SELECT * FROM elect LIMIT 15",db,index_col='index')
df
```

```python
df=pd.read_sql("SELECT DISTINCT country FROM elect",db)
df
```

```python
'Uzbekistan' in df['country'].values
```

```python
df=pd.read_sql("SELECT * FROM elect WHERE country='Uzbekistan' AND year>2010;",db)
df
```

```python
df=pd.read_sql("SELECT country,code,year,coal,gas,hydro FROM elect WHERE country='Uzbekistan' AND year>2010;",db)
df
```
